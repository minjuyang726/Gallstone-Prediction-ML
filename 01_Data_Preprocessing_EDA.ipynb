```python
#많은 부분에서 이상치를 발견한 행 제거
data = data.drop(23)
```

```python
data=data.drop(columns=['Obesity (%)'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

target_cols = ['Gallstone Status', 'Gallstone Status Int']
features = data.drop(columns=target_cols).select_dtypes(include=['number'])


# 상관행렬 계산
corr_matrix = features.corr()

# 히트맵 그리기
plt.figure(figsize=(15, 12))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    square=True,
    cbar_kws={"shrink": .8},
    annot_kws={"size": 5}
)

plt.xticks(rotation=45, ha='right')  # x축 라벨 45도 기울이기
plt.yticks(rotation=0)                # y축 라벨 수평 유지
plt.title('Feature Correlation Heatmap (excluding target)')
plt.tight_layout()
plt.show()
```
